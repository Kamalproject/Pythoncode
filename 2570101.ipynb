```python
#check number is prime or not
number=int(input("Enter a number"))
count=0
for i in range(1,number+1):
    if(number%i==0):
        count+=1
if(count==2):
    print("Number ",number," is a Prime Number")
elif(count==1):
    print("It is a Unique number")
else:
    print(number," is a composit number")
```

```python
#check weather it is Aramstrong or not
number=int(input("Enter a number"))
digit=len(str(number))
temp=number
digit_sum=0
while(temp>0):
    digits_count=temp%10
    digit_sum+=digits_count**digit
    temp//=10
if(number==digit_sum):
    print("Number ",number," is Aramstrong")
else:
    print("Number ",number," is not a Aramstrong")
```

```python
#Check the number is palindrom or not
number=int(input("Enetr a number"))
temp=number
reverse=0
while(temp>0):
    digit=temp%10
    reverse=reverse*10+digit
    temp//=10
if(reverse==number):
    print("Number ",number," is Palindrom")
else:
    print("Number ",number," is not a Palindrom")
```

```python
#Check wheater the number is perfect or not
number=int(input("Enter number"))
sum=0
for i in range(1,number):
    if(number%i==0):
        sum+=i
if(number==sum):
    print("Number",number,"is perfect number")
else:
    print("Number",number,"is not a perfect number")
```

```python
#Fibonacci series
number=int(input("Enter a number"))
count=0
num1=0
num2=1
while(count<number):
    print(num1,end=" ")
    num3=num1+num2
    num1=num2
    num2=num3
    count+=1
```

```python
#Create a list containing the mixed value of even, odd, two-digit and three-digit values. Filter 
#and generate separate list for all the above categories from the main list and display.

list=[2, 7, 12, 45, 8, 99, 105, 234, 56, 3, 678, 81]
even_list=[]
odd_list=[]
twodigit_list=[]
threedigit_list=[]
for i in list:
    if(i%2==0):
        even_list.append(i)      
    else:        
        odd_list.append(i)
for i in list:
    if(i>=10 and i<=99):
        twodigit_list.append(i)
    else:
        if(i>=100 and i<=700):
            threedigit_list.append(i)
print(even_list)
print(odd_list)
print(twodigit_list)
print(threedigit_list)
```
